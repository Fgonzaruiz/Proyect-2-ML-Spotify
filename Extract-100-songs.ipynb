```python
song_name =list()
song_artist =list()
song_popularity =list()
song_duration_ms =list()
acousticness =list()
danceability =list()
energy =list()
instrumentalness =list()
key =list()
liveness =list()
loudness =list()
audio_mode =list()
speechiness =list()
tempo =list()
time_signature =list()
audio_valence =list()

results = sp.search(q='year:2023', type='track', limit=50, offset=0)
tracks = results['tracks']['items']

for track in tracks:
    song_name.append(track['name'])
    song_artist.append(track['artists'][0]['name'])
    song_popularity.append(track['popularity'])

results = sp.search(q='year:2023', type='track', limit=50, offset=50)
tracks = results['tracks']['items']

for track in tracks:
    song_name.append(track['name'])
    song_artist.append(track['artists'][0]['name'])
    song_popularity.append(track['popularity'])

for i in range(len(song_name)):
    track_info = sp.search(q=f"{song_name[i]} {song_artist[i]}", type='track', limit=1)
    track_audio_features = sp.audio_features(track_info['tracks']['items'][0]['id'])[0]
    song_duration_ms.append(track_audio_features['duration_ms'])
    acousticness.append(track_audio_features['acousticness'])
    danceability.append(track_audio_features['danceability'])
    energy.append(track_audio_features['energy'])
    instrumentalness.append(track_audio_features['instrumentalness'])
    key.append(track_audio_features['key'])
    liveness.append(track_audio_features['liveness'])
    loudness.append(track_audio_features['loudness'])
    audio_mode.append(track_audio_features['mode'])
    speechiness.append(track_audio_features['speechiness'])
    tempo.append(track_audio_features['tempo'])
    time_signature.append(track_audio_features['time_signature'])
    audio_valence.append(track_audio_features['valence'])
    
# Create the DataFrame
df = pd.DataFrame({
    'song_name':song_name,
    'artist_name':song_artist,
    'song_popularity':song_popularity,
    'song_duration_ms':song_duration_ms,
    'acousticness':acousticness,
    'danceability':danceability,
    'energy':energy,
    'instumentalness':instrumentalness,
    'key':key,
    'liveness':liveness,
    'loudness':loudness,
    'audio_mode':audio_mode,
    'speechiness':speechiness,
    'tempo':tempo,
    'time_signature':time_signature,
    'audio_valence':audio_valence
})

df.to_csv('song_data_2023.csv', index=False)
```
